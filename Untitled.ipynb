```python
import pandas as pd
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
%%time 

df = pd.read_csv("movie_reviews.csv")
df.head()
```

```python
df2 = df[['movie_title','genre','movie_info']]
df2.head()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
Analyzer = SentimentIntensityAnalyzer()
```

```python
%%time
df2['Vader_scores']=df2['movie_info'].apply(Analyzer.polarity_scores)
df2.head()
```

```python
df2['Vader_score'] = df2['Vader_scores'].apply(lambda x: x['compound'])
df2 = df2[['movie_title','genre','movie_info','Vader_score']]
df2.head()
```

```python
def classify_sentiment(score):
    if score >= 0.05:
        return "Positive"
    elif score <= -0.05:
        return "Negative"
    else:
        return "Neutral"

df2["sentiment"] = df2["Vader_score"].apply(classify_sentiment)
```
